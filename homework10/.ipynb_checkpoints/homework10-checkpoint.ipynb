```python
import pandas as pd

df = pd.read_csv('solar_system.csv')

print("Actual columns in file:", df.columns.tolist())

df.columns = df.columns.str.strip().str.lower()

cols_to_fix = ['distance', 'mass', 'radius'] 

for col in cols_to_fix:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(r'[^0-9.]', '', regex=True), errors='coerce')
    else:
        print(f"Warning: Column '{col}' still not found!")

df = df.dropna()
print(df.head())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy import units as u
from astropy import constants as const

df = pd.read_csv('solar_system.csv')

print(f"Shape before: {df.shape}") # (Rows, Cols)
df = df.set_index("Attribute").T
df.index.name = "Planet"
df.reset_index(inplace=True)
df.columns.name = None
print(f"Shape after: {df.shape}")

print(df.dtypes)

for col in df.columns:
    print(f"{col}: {df[col].apply(type).unique()}")

for col in df.columns:
    if col not in ["Planet", "Ring System?", "Global Magnetic Field?"]:
        df[col] = pd.to_numeric(df[col], errors="coerce")

print(df.dtypes)
```

```python
def attach_units(df, col_name, unit):
    """
    Applies an astropy unit to a column and renames it.
    """
    values_with_units = df[col_name].values * unit
    
    new_name = f"{col_name} ({unit})"
    df = df.rename(columns={col_name: new_name})
    
    return values_with_units, df
    
mass_q, df = attach_units(df, "Mass", u.kg)
print(df[["Planet", "Mass (kg)"]].head())
```

```python
# 2.3.1 Semi-Major Axis
# Assuming columns are named 'Perihelion' and 'Aphelion'
df['Semi-Major Axis'] = (df['Perihelion'] + df['Aphelion']) / 2
# Insert it after Aphelion (find index of Aphelion and add 1)
idx = df.columns.get_loc("Aphelion") + 1
col_data = df.pop("Semi-Major Axis")
df.insert(idx, "Semi-Major Axis", col_data)
print(df.columns)

# 2.3.2 Convert Orbital Period (Days -> Years)
# Assuming the column is 'Orbital Period'
orbital_days = df['Orbital Period'].values * u.day
df['Orbital Period'] = orbital_days.to(u.year).value
df = df.rename(columns={'Orbital Period': 'Orbital Period (years)'})

# F-string for favorite planet (e.g., Earth at index 2)
fav_planet_period = df.loc[df['Planet'] == 'Earth', 'Orbital Period (years)'].values[0]
print(f"Earth's orbital period: {fav_planet_period:.4f} years")

# 2.3.3 Distance Columns (km -> AU)
print(f"One AU in km: {const.au.to(u.km)}")

dist_cols = ['Diameter', 'Distance from Sun', 'Perihelion', 'Aphelion', 'Semi-Major Axis']
for col in dist_cols:
    if col in df.columns:
        km_val = df[col].values * u.km
        df[col] = km_val.to(u.au).value
        df = df.rename(columns={col: f"{col} (AU)"})
```
