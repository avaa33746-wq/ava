```python
import numpy as np
lst = [20, 27, 9, 12, 31, 17, 4, 50, 39, 35]
arr = np.array(lst)
Multiply by 4: print(arr * 4) → [80, 108, 36, 48, 124, 68, 16, 200, 156, 140].
Mean, Median, & Std Dev:
Mean: np.mean(arr) is 24.4.
Median: np.median(arr) is 23.5.
Standard Deviation: np.std(arr) is ~13.83.
Argmin & Argmax:
min_idx = np.argmin(arr) (Index 6)
max_idx = np.argmax(arr) (Index 7)
Values from indices: arr[min_idx] is 4; arr[max_idx] is 50.
Min & Max comparison: np.min(arr) is 4 and np.max(arr) is 50. Yes, they match the results from the previous step.
Boolean indexing (> 20): print(arr[arr > 20]) → [27, 31, 50, 39, 35].
```

```python
matrix = np.array([[5, 12, 17, 9], [8, 21, 14, 3], [6, 13, 19, 11]])
Shape: matrix.shape is (3, 4). There are 3 rows and 4 columns.
Third column: matrix[:, 2] → [17, 14, 19].
Second row: matrix[1, :] → [8, 21, 14, 3].
Multiply by 2: print(matrix * 2) doubles every element.
Means:
Column Mean: np.mean(matrix, axis=0) → [6.33, 15.33, 16.67, 7.67].
Overall Mean: np.mean(matrix) is 11.5.
Max/Min stats:
Max in each row: np.max(matrix, axis=1) → [17, 21, 19].
Min in each column: np.min(matrix, axis=0) → [5, 12, 14, 3].
Flatten: flat = matrix.flatten(). New shape is (12,). The array was collapsed into a single dimension.
Flattened indices: np.argmin(matrix) is 7; np.argmax(matrix) is 5.
Unravel indices: np.unravel_index(7, matrix.shape) → (1, 3); np.unravel_index(5, matrix.shape) → (1, 1).
Boolean indexing (> 15): matrix[matrix > 15] → [17, 21, 19].
Sort each row: np.sort(matrix, axis=1)
```

```python
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def get_prime_rows(arr):
    # Vectorize the prime check function to apply it to the array
    v_is_prime = np.vectorize(is_prime)
    prime_mask = v_is_prime(arr)
    
    # Use .any(axis=1) to find rows where at least one True exists
    return arr[prime_mask.any(axis=1)]

# Test with your array
arr = np.array([
    [2, 3, 5],
    [14, 6, 8],
    [11, 13, 17],
    [7, 10, 13]
])

print(get_prime_rows(arr))
```

```python
def checkerboard():
    # 5.1: Create 8x8 array of all zeros
    board = np.zeros((8, 8), dtype=int)
    
    # 5.2: For odd-numbered rows (index 0, 2, 4, 6), 
    # start at index 0 and fill every other column with 1
    board[0::2, 0::2] = 1
    
    # 5.3: For even-numbered rows (index 1, 3, 5, 7), 
    # start at index 1 and fill every other column with 1
    board[1::2, 1::2] = 1
    
    return board

print(checkerboard())
```

```python
def reverse_checkerboard():
    board = np.zeros((8, 8), dtype=int)
    
    # Fill even columns for even rows (index 0, 2, 4, 6)
    # This places a '1' at [0,1], [0,3], etc., making the first element [0,0] a '0'
    board[0::2, 1::2] = 1
    
    # Fill odd columns for odd rows (index 1, 3, 5, 7)
    board[1::2, 0::2] = 1
    
    return board

print(reverse_checkerboard())
```

```python
def expansion(arr, n):
    # Create the gap string (n spaces)
    gap = " " * n
    
    # Helper to join letters with the gap
    def expand_word(word):
        return gap.join(list(word))
    
    # Apply to the entire array at once
    v_expand = np.vectorize(expand_word)
    return v_expand(arr)

universe = np.array(['galaxy', 'clusters'])
print(expansion(universe, 1))
```

```python
import numpy as np  
def second_brightest(stars):
    # Sort each row (ascending order)
    sorted_stars = np.sort(stars, axis=1)
    
    # Pick the second to last element from each row [-2]
    return sorted_stars[:, 1]

# Test with your example
np.random.seed(123)
stars = np.random.randint(500, 2000, (5, 5))
print(second_brightest(stars))
# Output: [1789, 1678, 1654, 1784, 1756]
```
