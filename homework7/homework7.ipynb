```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define the sine wave function
def make_sine_wave(x, A, w):
    return A * np.sin(w * x)

# 2. Create the x array from 0 to 2pi with 1000 points
x = np.linspace(0, 2 * np.pi, 1000)

# 3. Define amplitude and frequency lists
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]

# Define linestyles for the bonus part
linestyles = ['-', '--', ':', '-.', (0, (3, 5, 1, 5))]

# 4. Create three side-by-side subplots
fig, axes = plt.subplots(1, 3, figsize=(25, 5))

# 5 & 6. Loop through both arrays using zip() and plot
for i, (amp, freq) in enumerate(zip(amplitude, frequency)):
    ls = linestyles[i % len(linestyles)] # Cycle through linestyles
    
    # Subplot 1: Vary only amplitude (set w to 1)
    y1 = make_sine_wave(x, amp, 1)
    axes[0].plot(x, y1, linestyle=ls, label=f"A={amp}, ω=1")
    
    # Subplot 2: Vary only frequency (set A to 1)
    y2 = make_sine_wave(x, 1, freq)
    axes[1].plot(x, y2, linestyle=ls, label=f"A=1, ω={freq}")
    
    # Subplot 3: Vary both
    y3 = make_sine_wave(x, amp, freq)
    axes[2].plot(x, y3, linestyle=ls, label=f"A={amp}, ω={freq}")

# 7. Add titles, labels, and legends with increased font size
titles = ["Varying Amplitude", "Varying Frequency", "Varying Both"]
for i, ax in enumerate(axes):
    ax.set_title(titles[i], fontsize=18)
    ax.set_xlabel("x values", fontsize=14)
    ax.set_ylabel("y = A sin(ωx)", fontsize=14)
    ax.legend(fontsize=10)

# 8. Display the plots
plt.tight_layout()
plt.show()
```
