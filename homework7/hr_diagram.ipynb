```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Load the data
# (Ensure star_data.txt is in the same folder as this notebook)
data = np.loadtxt("star_data.txt", skiprows=1)

# 2. Define the variables using array slicing
magnitude = data[:, 0]
temperature = data[:, 1]

# 3. Create log variables
log_magnitude = np.log10(magnitude)
log_temperature = np.log10(temperature)

# 4. Create the figure and plot
plt.figure(figsize=(8, 6))

# Bonus: Scatter plot with a colormap
plt.scatter(log_temperature, log_magnitude, c=temperature, cmap="RdYlBu", s=10, label="Stars")

# 5. Invert the axes
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

# 6. Highlight specific stars
# Brightest
plt.scatter(log_temperature[np.argmin(magnitude)], np.log10(np.min(magnitude)), 
            color="black", marker="*", s=300, label="Brightest Star")
# Dimmest
plt.scatter(log_temperature[np.argmax(magnitude)], np.log10(np.max(magnitude)), 
            color="purple", marker="*", s=300, label="Dimmest Star")
# Hottest
plt.scatter(np.log10(np.max(temperature)), log_magnitude[np.argmax(temperature)], 
            color="green", marker="*", s=300, label="Hottest Star")
# Coldest
plt.scatter(np.log10(np.min(temperature)), log_magnitude[np.argmin(temperature)], 
            color="orange", marker="*", s=300, label="Coldest Star")

# 7. Add Labels and Title
plt.title("Hertzsprung-Russell Diagram")
plt.xlabel("Temperature (log K)")
plt.ylabel("Magnitude")
plt.legend()

plt.show()
```
