```python
import pandas as pd
```

```python
# The "../" tells Python to go up one folder level before looking for the data
df = pd.read_csv('../../course_assignments/data/state_data.csv')

# Verify it loaded by checking the first few rows
df.head()
print(df.shape)
df.head()
df.tail()
median_lat = df['latitude'].median()
median_long = df['longitude'].median()

print(f"Median Latitude: {median_lat}")
print(f"Median Longitude: {median_long}")

df['lat_dist'] = (df['latitude'] - median_lat).abs()

# Check the new column
df.head()

# Create a new column for absolute distance to median longitude
df['long_dist'] = (df['longitude'] - median_long).abs()

# The output will show (rows, columns)
print(df.shape)

# Latitude Calculations
closest_lat_row = df.loc[df['lat_dist'].idxmin()]
farthest_lat_row = df.loc[df['lat_dist'].idxmax()]

# Longitude Calculations
closest_long_row = df.loc[df['long_dist'].idxmin()]
farthest_long_row = df.loc[df['long_dist'].idxmax()]

# Printing with rounding to two decimal places
print(f"The state closest to the median latitude is {closest_lat_row['state']} by {closest_lat_row['lat_dist']:.2f} degrees.")
print(f"The state farthest from the median latitude is {farthest_lat_row['state']} by {farthest_lat_row['lat_dist']:.2f} degrees.")
print(f"The state closest to the median longitude is {closest_long_row['state']} by {closest_long_row['long_dist']:.2f} degrees.")
print(f"The state farthest from the median longitude is {farthest_long_row['state']} by {farthest_long_row['long_dist']:.2f} degrees.")
```

```python
# Returns the 3 rows with the largest values in the 'lat_dist' column
top_3_lat_dist = df.nlargest(3, 'lat_dist')
print(top_3_lat_dist)
```

```python
# Summarizing the distance statistics for latitude and longitude using agg
dist_summary = df.agg({
    "lat_dist": ["min", "max", "mean", "std"],
    "long_dist": ["min", "max", "mean", "std"]
})
print(dist_summary)

Explain where you found it: I found this function under the API Reference tab of the Pandas documentation, listed within the "DataFrame" section under "Function application, GroupBy & window."
What the function does & Why it's helpful:
What it does: The .agg() function allows you to run multiple different mathematical operations (like mean, min, or standard deviation) on specific columns all at once.
When it is helpful: In this project, while nlargest helped me find specific outlier states, .agg() is helpful for seeing the overall distribution of my distance calculations. It allows me to see the average distance from the median and the total spread (standard deviation) of the states in one clean table, rather than running df['lat_dist'].mean(), df['lat_dist'].max(), etc., in separate lines.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset (picked from the seaborn-data repo)
df_penguins = sns.load_dataset("penguins")

# Drop all columns with NaN values
df_penguins = df_penguins.dropna(axis=1) 

# Note: If dropping columns removes too much, 
# you can use df_penguins.dropna() to drop rows instead.

Analyzing the Dataset
Shape: df_penguins.shape (returns rows, columns).
Rows/Columns: Use len(df_penguins) for rows and len(df_penguins.columns) for columns.
Head/Tail: df_penguins.head() and df_penguins.tail().
Columns:
species: Type of penguin (Adelie, Gentoo).
island: Where they were found (Torgersen, Biscoe, Dream).
bill_length_mm: Length of the penguin's beak.
bill_depth_mm: Thickness of the penguin's beak.
flipper_length_mm: Length of their flipper/wing.
body_mass_g: Weight in grams.

# Create 2x2 grid
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# 1. Top Left: Scatter (Bill Length vs Bill Depth)
sns.scatterplot(data=df_penguins, x="bill_length_mm", y="bill_depth_mm", 
                hue="species", ax=axes[0, 0], palette="viridis")
axes[0, 0].set_title("Beak Dimensions by Species")
axes[0, 0].set_xlabel("Bill Length (mm)")
axes[0, 0].set_ylabel("Bill Depth (mm)")
axes[0, 0].legend(title="Species")

# 2. Top Right: Scatter (Flipper Length vs Body Mass)
sns.scatterplot(data=df_penguins, x="flipper_length_mm", y="body_mass_g", 
                color="orange", ax=axes[0, 1])
axes[0, 1].set_title("Flipper Length vs. Weight")
axes[0, 1].set_xlabel("Flipper Length (mm)")
axes[0, 1].set_ylabel("Body Mass (g)")

# 3. Bottom Left: Histogram (Body Mass)
sns.histplot(data=df_penguins, x="body_mass_g", color="teal", kde=True, ax=axes[1, 0])
axes[1, 0].set_title("Distribution of Penguin Weight")
axes[1, 0].set_xlabel("Body Mass (g)")
axes[1, 0].set_ylabel("Count")

# 4. Bottom Right: Boxplot (Flipper Length by Island)
sns.boxplot(data=df_penguins, x="island", y="flipper_length_mm", palette="Set2", ax=axes[1, 1])
axes[1, 1].set_title("Flipper Length across Islands")
axes[1, 1].set_xlabel("Island Name")
axes[1, 1].set_ylabel("Flipper Length (mm)")

plt.show()

Explanation Examples
Top Left: Represents the relationship between beak length and depth. I chose these because different species often evolve different beak shapes for food. I learned that Adelie penguins have shorter, deeper bills than Gentoos.
Bottom Right: Represents flipper length variation across islands. I chose this to see if geography affects size. I learned that penguins on Biscoe island tend to have significantly larger flippers.
```
